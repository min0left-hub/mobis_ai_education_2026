## 1-A. нҢҢмқҙмҚ¬ мҪ”л”©мңјлЎң нҷ•мқён•ҳлҠ” м„ нҳ• нҡҢк·Җ

###1. нҷҳкІҪ мӨҖл№„

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
```

### 2. лҚ°мқҙн„° мӨҖл№„

```python
# кіөл¶Җ мӢңк°„ XмҷҖ м„ұм Ғ yмқҳ л„ҳнҢҢмқҙ л°°м—ҙ мғқм„ұ
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
# лҚ°мқҙн„°мқҳ л¶„нҸ¬лҘј к·ёлһҳн”„ н”ҢлЎҜнҢ…
plt.scatter(x, y)
plt.show()
```

### LINEAR REGRESSION мӢқ мҙҲкё° м„Өм •

```python
# кё°мҡёкё° a, м ҲнҺё b к°’ мҙҲкё°нҷ”
a = 0
b = 0

#н•ҷмҠөлҘ  м„Өм •
lr=0.001

#л°ҳліөнҡҹмҲҳ
epochs=200
```

### кІҪмӮ¬ н•ҳк°•лІ•

```python
# x к°’мқҙ мҙқ лӘҮ к°ңмқём§Җ м…үлӢҲлӢӨ.
n=len(x)

# кІҪмӮ¬ н•ҳк°•лІ•мқ„ мӢңмһ‘н•©лӢҲлӢӨ.
for i in range(epochs):                  # м—җнҸ¬нҒ¬ мҲҳ л§ҢнҒј л°ҳліө

    y_pred = a * x + b                   # мҳҲмёЎ к°’мқ„ кө¬н•ҳлҠ” мӢқмһ…лӢҲлӢӨ.
    error = y - y_pred                   # мӢӨм ң к°’кіј л№„көҗн•ң мҳӨм°ЁлҘј errorлЎң лҶ“мҠөлӢҲлӢӨ.

    a_diff = (2/n) * sum(-x * (error))   # мҳӨм°Ё н•ЁмҲҳлҘј aлЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.
    b_diff = (2/n) * sum(-(error))       # мҳӨм°Ё н•ЁмҲҳлҘј bлЎң нҺёлҜёл¶„н•ң к°’мһ…лӢҲлӢӨ.

    a = a - lr * a_diff     # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ a к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.
    b = b - lr * b_diff     # н•ҷмҠөлҘ мқ„ кіұн•ҙ кё°мЎҙмқҳ b к°’мқ„ м—…лҚ°мқҙнҠён•©лӢҲлӢӨ.

    if i % 10 == 0:        # 100лІҲ л°ҳліөлҗ  л•Ңл§ҲлӢӨ нҳ„мһ¬мқҳ a к°’, b к°’мқ„ м¶ңл Ҙн•©лӢҲлӢӨ.
        #print("epoch=%.f, кё°мҡёкё°=%.04f, м ҲнҺё=%.04f" % (i, a, b))
        # м•һм„ң кө¬н•ң мөңмў… a к°’мқ„ кё°мҡёкё°, b к°’мқ„ y м ҲнҺём—җ лҢҖмһ…н•ҳм—¬ к·ёлһҳн”„лҘј к·ёлҰҪлӢҲлӢӨ.
        y_pred = a * x + b
        # к·ёлһҳн”„ м¶ңл Ҙ
        plt.scatter(x, y)
        plt.plot(x, y_pred,'r')
        plt.show()
        display.clear_output(wait=True)
```

```python
print("linear regression мӢқ : y = %.02fx+%.02f " %(a,b))
```

### м җмҲҳ мҳҲмёЎн•ҳкё°

```python
hour = 7
prediction = a*hour + b

print("%.fмӢңк°„мқ„ кіөл¶Җн•  кІҪмҡ°мқҳ мҳҲмғҒ м җмҲҳлҠ” %.02fм җмһ…лӢҲлӢӨ" %(hour, prediction))
```

## лӘЁлҚёмқ„ м ҖмһҘн•ҙм„ң м җмҲҳлҘј мҳҲмёЎн•ҳкё°

```python
# нҷҳкІҪ мӨҖл№„
import numpy as np
import matplotlib.pyplot as plt

# н…җм„ңн”ҢлЎңмқҳ мјҖлқјмҠӨ APIм—җм„ң н•„мҡ”н•ң н•ЁмҲҳл“Өмқ„ л¶Ҳлҹ¬мҳөлӢҲлӢӨ.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
# лҚ°мқҙн„° мӨҖл№„
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
# лӘЁлҚё м•„нӮӨн…ҚмІҳ м„Өм •
model = Sequential()

# м¶ңл Ҙ к°’, мһ…л Ҙ ліҖмҲҳ, л¶„м„қ л°©лІ•м—җ л§һкІҢлҒ” лӘЁлҚёмқ„ м„Өм •н•©лӢҲлӢӨ.
model.add(Dense(1, input_dim=1, activation='linear'))
optimizer_0p3 = SGD(learning_rate=0.001)

# мҳӨм°Ё мҲҳм •мқ„ мң„н•ҙ кІҪмӮ¬ н•ҳк°•лІ•(sgd)мқ„, мҳӨм°Ёмқҳ м •лҸ„лҘј нҢҗлӢЁн•ҳкё° мң„н•ҙ нҸүк·  м ңкіұ мҳӨм°Ё(mse)лҘј мӮ¬мҡ©н•©лӢҲлӢӨ.
model.compile(optimizer=optimizer_0p3, loss='mse')

```

```python
# лӘЁлҚё н•ҷмҠө
# мҳӨм°ЁлҘј мөңмҶҢнҷ”н•ҳлҠ” кіјм •мқ„ 200лІҲ л°ҳліөн•©лӢҲлӢӨ.
model.fit(x, y, epochs=200)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x),'r')    # мҳҲмёЎ кІ°кіјлҘј к·ёлһҳн”„лЎң лӮҳнғҖлғ…лӢҲлӢӨ.
plt.show()
```

```python
# лӘЁлҚё н…ҢмҠӨнҠё
# мһ„мқҳмқҳ мӢңк°„мқ„ м§‘м–ҙл„Јм–ҙ м җмҲҳлҘј мҳҲмёЎн•ҳлҠ” лӘЁлҚёмқ„ н…ҢмҠӨнҠён•ҙ ліҙкІ мҠөлӢҲлӢӨ.

hour = 7
prediction = model.predict(np.array([hour]))

print("%.fмӢңк°„мқ„ кіөл¶Җн•  кІҪмҡ°мқҳ мҳҲмғҒ м җмҲҳлҠ” %.02fм җмһ…лӢҲлӢӨ" % (hour, prediction))
```
